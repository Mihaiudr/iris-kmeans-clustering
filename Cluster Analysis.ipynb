# Species Segmentation with Cluster Analysis

The Iris flower dataset is one of the most popular ones for machine learning. https://en.wikipedia.org/wiki/Iris_flower_data_set

There are 4 features: sepal length, sepal width, petal length, and petal width.

I will want to create clusters of the Iris flower to predict the reality.

## Import the relevant libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from sklearn.cluster import KMeans
```

## Load the data

Load data from the csv file: <i> 'iris_dataset.csv'</i>.

```python
# Load the data
data = pd.read_csv('iris_dataset.csv')
# Check the data
data
```

## Plot the data

For this exercise, I will try to cluster the iris flowers by the shape of their sepal. 

<i> I used the 'sepal_length' and 'sepal_width' variables.</i> 

```python
#scatter plot based on two corresponding features (sepal_length and sepal_width; OR petal_length and petal_width)
plt.scatter(data['sepal_length'],data['sepal_width'])
plt.xlabel('Lenght of sepal')
plt.ylabel('Width of sepal')
plt.show()
```

## Clustering (unscaled data)

```python
# created a variable which will contain the data for the clustering
x = data.copy()
# created a k-means object with 2 clusters
kmeans = KMeans(2)
# fit the data
kmeans.fit(x)
```

```python
# created a copy of data, so we can see the clusters next to the original data
clusters = data.copy()
# cluster prediction for each observation
clusters['cluster_pred']=kmeans.fit_predict(x)
```

```python
# scatter plot based on two corresponding features (sepal_length and sepal_width; OR petal_length and petal_width)
plt.scatter(clusters['sepal_length'], clusters['sepal_width'], c= clusters ['cluster_pred'], cmap = 'rainbow')
```

## Standardize the variables

I will import and use the <i> scale </i> method from sklearn to standardize the data. 

```python
# import some preprocessing module
from sklearn import preprocessing

# scale the data for better results
x_scaled = preprocessing.scale(data)
x_scaled
```

## Clustering (scaled data)

```python
# create a k-means object with 2 clusters
kmeans_scaled = KMeans(2)
# fit the data
kmeans_scaled.fit(x_scaled)
```

```python
# create a copy of data, so we can see the clusters next to the original data
clusters_scaled = data.copy()
# predict the cluster for each observation
clusters_scaled['cluster_pred']=kmeans_scaled.fit_predict(x_scaled)
```

```python
# create a scatter plot based on two corresponding features (sepal_length and sepal_width)
plt.scatter(clusters_scaled['sepal_length'], clusters_scaled['sepal_width'], c= clusters_scaled ['cluster_pred'], cmap = 'rainbow')
```

Looks like the two solutions are identical. That is because the original features have very similar scales to start with.

## The Elbow Method

### WCSS

```python
wcss = []
# 'cl_num' is a variable that keeps track the highest number of clusters we want to use the WCSS method for. 
# We have it set at 10 right now, but it is completely arbitrary.
cl_num = 10
for i in range (1,cl_num):
    kmeans= KMeans(i)
    kmeans.fit(x_scaled)
    wcss_iter = kmeans.inertia_
    wcss.append(wcss_iter)
wcss
```

### The Elbow Method

```python
number_clusters = range(1,cl_num)
plt.plot(number_clusters, wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('Within-cluster Sum of Squares')
```

### Silhouette score

```python
from sklearn.metrics import silhouette_score

for k in [2,3,5]:
    kmeans = KMeans(k, random_state=1)
    label = kmeans.fit_predict(x_scaled)
    score = silhouette_score(x_scaled, label)
    print(f"Silhouette score for k={k}: {score:.4f}")
```

The silhouette analysis indicated that k=2 provides the strongest cluster separation, reflecting the fact that Setosa is clearly distinct while Versicolor and Virginica overlap substantially in feature space.

```python
# PCA

from sklearn.decomposition import PCA

pca = PCA(n_components=2)
x_pca = pca.fit_transform(x_scaled)

plt.figure()
plt.scatter(x_pca[:,0],x_pca[:,1],c=clusters_scaled['clusters3'], cmap='viridis')
plt.title('K-means clusters k=3 in PCA space')
plt.xlabel('PCA component 1')
plt.ylabel('PCA component 2')
plt.colorbar()
plt.show()
```

## Compare the solutions to the original iris dataset

The original (full) iris data is located in <i>iris_with_answers.csv</i>. I will load the csv, plot the data and compare it with your solution. 

Obviously there are only 3 types, because that's the original (truthful) iris dataset.

The 2-cluster solution seemed good, but in real life the iris dataset has 3 SPECIES (a 3-cluster solution). Therefore, clustering cannot be trusted at all times. Sometimes it seems like x clusters are a good solution, but in real life, there are more (or less).

```python
raw_data=pd.read_csv('iris_with_answers.csv')
```

```python
raw_data['species'].unique()
```

```python
raw_data['species_bin']=raw_data['species'].map({'setosa':0,'versicolor':1,'virginica':2})
raw_data
```

# Comparison between the original species and the clustered species

```python
fig, axes = plt.subplots(ncols=2, figsize=(12,5))

#Real data

axes[0].scatter(raw_data['sepal_length'], raw_data['sepal_width'], c=raw_data['species_bin'], cmap='rainbow')
axes[0].set_title("Real species")
axes[0].set_xlabel('Sepal Length')
axes[0].set_ylabel('Sepal Width')

#Clustered data

axes[1].scatter(clusters_scaled['sepal_length'], clusters_scaled['sepal_width'], c=clusters_scaled['clusters3'], cmap='rainbow')
axes[1].set_title("Clusters Kmeans")
axes[1].set_xlabel("Sepal Length")
axes[1].set_ylabel("Sepal Width")
```

```python
pd.crosstab(raw_data['species'], clusters_scaled['clusters3'])
```

The crosstab comparison shows that K-means perfectly identifies Setosa (50/50 correctly grouped), which is the most distinct species. Versicolor and Virginica overlap across clusters 1 and 2, reflecting their natural similarity and the known difficulty of separating them using unsupervised methods.

### Hungarian algorithm

```python
from sklearn.metrics import accuracy_score
from scipy.optimize import linear_sum_assignment

true_labels = raw_data['species_bin']
pred_labels = clusters_scaled['clusters3']

# Build confusion matrix
conf_matrix = pd.crosstab(true_labels, pred_labels).values

# Hungarian algorithm to best match cluster label
row_ind, col_ind = linear_sum_assignment(conf_matrix.max() - conf_matrix)

accuracy = conf_matrix[row_ind, col_ind].sum() / conf_matrix.sum()
print("Clustering accuracy:", round(accuracy, 4))
```

### GMM  Clustering Algorithm

In addition to K-Means, I applied a Gaussian Mixture Model (GMM) with three components to cluster the Iris dataset. GMM is a probabilistic clustering method that models each cluster as a Gaussian distribution and can better handle overlapping classes. I evaluated GMM using silhouette score and a crosstab/confusion matrix against the true species labels, and compared its performance with K-Means.

```python
from sklearn.mixture import GaussianMixture

gmm = GaussianMixture(n_components=3, random_state=1)
x_gmm = gmm.fit_predict(x_scaled)
gmm_clusters=clusters_scaled.copy()
gmm_clusters['GMM'] = x_gmm
gmm_clusters
```

```python
plt.scatter(gmm_clusters['sepal_length'], gmm_clusters['sepal_width'], c=gmm_clusters['GMM'], cmap='rainbow')
```

```python
# GMM vs Real Species

pd.crosstab(raw_data['species'], gmm_clusters['GMM'], rownames=['Actual'], colnames=['GMM clusters'])
```

The Gaussian Mixture Model (GMM) correctly identified the Setosa and Virginica classes with 100% purity, assigning all samples to single clusters. GMM struggled slightly with Versicolor, which naturally overlaps with Virginica in feature space, leading to a small number of misclustered samples. Overall, GMM produced more biologically meaningful clusters than K-Means, reflecting its flexibility in modeling overlapping distributions.

```python
# Hungarian algoritm for GMM

true_labels = raw_data['species_bin']
pred_labels_GMM = gmm_clusters['GMM']

# Build confusion matrix
conf_matrix = pd.crosstab(true_labels, pred_labels_GMM).values

# Hungarian algorithm to best match cluster label
row_ind, col_ind = linear_sum_assignment(conf_matrix.max() - conf_matrix)

accuracy = conf_matrix[row_ind, col_ind].sum() / conf_matrix.sum()
print("Clustering accuracy:", round(accuracy, 4))
```

### Comparison between GMM clustering and real species

```python
fig, axes = plt.subplots(ncols=2, figsize=(12,5))

#Real data

axes[0].scatter(raw_data['sepal_length'], raw_data['sepal_width'], c=raw_data['species_bin'], cmap='rainbow')
axes[0].set_title("Real species")
axes[0].set_xlabel('Sepal Length')
axes[0].set_ylabel('Sepal Width')

#Clustered data with GMM

axes[1].scatter(gmm_clusters['sepal_length'], gmm_clusters['sepal_width'], c=gmm_clusters['GMM'], cmap='rainbow')
axes[1].set_title("Clusters GMM")
axes[1].set_xlabel("Sepal Length")
axes[1].set_ylabel("Sepal Width")
```

### Conclusion

The Gaussian Mixture Model achieved a clustering accuracy of 96.67%, meaning it correctly grouped almost all samples into clusters that align with the real Iris species. GMM perfectly separated Setosa and Virginica, while only a small portion of Versicolor samples were confused with the overlapping Virginica class. Compared to K-Means, GMM produced clusters that more closely match the natural distribution of the species.
